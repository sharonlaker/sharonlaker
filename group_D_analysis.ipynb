Defining a function that removes space at the end of a string

```python
# we define a function to clean and remove spaces at the end
def remove_endspace(data):
    
    #checking if type of argument is a list
    if type(data) is list:
        newlist = []
        for word in data:
            
            #applied some recursion
            newlist.append(remove_endspace(word))
        return newlist
    
    elif type(data) is str:
        
        #checking if string ends with a space
        if data.endswith(" "):
            
            #return the string without that end space
            return data[:(len(data)-1)]
        else:
            return data
    else:
        return data
```

#### This function is to be used to group all the data accordding to regions

```python
"""
Note:
This function makes use of our earlier defined dictionary 'districts_upper' that contains districts
grouped by regions.
"""
def regional_grouper(data):
    
    #dictionary to contain dataFrames of regions
    regional_dict ={"n":[],
                    "e":[],
                    "w":[],
                    "c":[]
                   }
    
    #Resetting index of the passed dataFrame to minimize errors
    data.reset_index(inplace=True)
    data.drop(columns="index",inplace=True)
    
    #looping through the indices of the passed dataFrame
    for index in data.index:
        
        #looping through the districts dictionary to obtain regions
        for region in districts_upper:
            
            #checking for district in the district dictionary
            if(data.iloc[index]["DISTRICT"] in districts_upper[region]):
                
                #categorizing the district into its region
                if region is "eastern":
                    regional_dict["e"].append(data.iloc[index])
                elif region is "western":
                    regional_dict["w"].append(data.iloc[index])
                elif region is "central":
                    regional_dict["c"].append(data.iloc[index])
                elif region is "northern":
                    regional_dict["n"].append(data.iloc[index])
                    
    #Returning our grouped data according to regions
    return regional_dict

```

#### This function will be used to plot only piecharts of regional data

```python
#This function will help us draw pie charts of our grouped regional data
#It accepts the data to be plotted as the parameter,labels,figsize and the shadow

def pie_chart(data,labels,figsize=(12,12),shadow=True,title=""):
    
    #Setting our axes and figure objects and setting the figsize
    fig,ax = plt.subplots(figsize=(12,12))
    plt.rcParams['font.size']=19
    
    #setting the title for the pie-chart and increasing the fontsize to 15
    ax.set_title(title,fontsize=15)
    
    #Exploding our piechart like a chopped pizza
    explode = (0.04,0.04,0.04,0.04)
    
    #initiating our pie-chart with the provided data
    ax.pie(data,
           labels=labels,
           explode=explode,
           shadow=shadow,
           autopct="%1.1f%%"  
          )
    ax.axis("equal")
    
    #Showing the pie chart
    plt.show()
    
    #closing the plt object to save resources
    plt.close()

#NB
#This function makes use of the matplotlib library
#Hence should be installed on your PC
```
